Analysis Part


```python
team_goals = data.groupby('Team')['Goal'].sum().sort_values(ascending=False).head(10)
print("Total Goals by Team:\n", team_goals)
```

```python
print(data.groupby('Team')['Red Cards'].sum().sort_values(ascending=False).head(10))
print(data.groupby('Team')['Yellow Cards'].sum().sort_values(ascending=False).head(10))
```

```python
data['Total Cards'] = data['Red Cards'] + data['Yellow Cards']
```

```python
data.groupby('Team')['Total Cards'].sum().sort_values(ascending=False).head(10)
```

```python
data.groupby("Team")[["Inside Penalty Area","Outside Penalty Area"]].sum().head(10)
```

```python
data.groupby("Team")["Penalties Scored"].sum().sort_values(ascending=False).head(10)
```

```python
numeric_data = data.select_dtypes(include=np.number)
plt.figure(figsize=(18, 12))
sns.heatmap(numeric_data.corr(),annot=True,fmt=".2f")
```
